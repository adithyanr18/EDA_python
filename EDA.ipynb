```python
#2.is there a signifcant variance in profit across different product subcategories?
import pandas as pd

# Step 1: Create the Profit/Sales ratio
data['Profit_Sales_Ratio'] = data['Profit'] / data['Sales']

# Step 2: Group by Sub-Category
grouped = data.groupby('Sub-Category')

# Step 3: Calculate Std Dev and Covariance for each group
results = []

for name, group in grouped:
    std_dev = group['Profit_Sales_Ratio'].std()
    cov = group[['Profit', 'Sales']].cov().loc['Profit', 'Sales']
    results.append({
        'Sub-Category': name,
        'Std_Dev_Profit/Sales': std_dev,
        'Cov_Profit_Sales': cov
    })

# Step 4: Create summary DataFrame
summary_df = pd.DataFrame(results).sort_values(by='Std_Dev_Profit/Sales', ascending=False)
print(summary_df)
```

```python
#3.use frequency table tables for returned and then calculate mean,std,variance of sales between refunded and non-refunded orders and compare them by visualizing side by side boxplots

data.head(5)
# Create a frequency table for 'Returned'
frequency_table = data['Returned'].value_counts()
print(frequency_table)

# Calculate and compare mean, std, and variance of 'Sales' between returned and non-returned orders
returned_sales = data[data['Returned'] == 'Yes']['Sales']
mean_returned = returned_sales.mean()
std_returned = returned_sales.std() 
var_returned = returned_sales.var()


non_returned_sales = data[data['Returned'] == 'Not']['Sales']
mean_non_returned = non_returned_sales.mean()  
std_non_returned = non_returned_sales.std()
var_non_returned = non_returned_sales.var()

#mean, std, var of returned and non-returned orders
print(f"returned - Mean: {mean_returned}, Std: {std_returned}, Var: {var_returned}")
print(f"non-returned - Mean: {mean_non_returned}, Std: {std_non_returned}, Var: {var_non_returned}")

# Create side-by-side boxplots for returned and non-returned orders
plt.figure(figsize=(12, 6))
sns.boxplot(x='Returned', y='Sales', data=data)
plt.title('Boxplot of Sales by Returned Status')
plt.xlabel('Returned Status')   
plt.ylabel('Sales')
plt.show()
```

```python
#4.for each retail sales person calculate correlation and covaraince between discount and profit
data['Discount_Profit_Correlation'] = data.groupby('Retail Sales People')['Discount'].transform(lambda x: x.corr(data['Profit']))
data['Discount_Profit_Covariance'] = data.groupby('Retail Sales People')['Discount'].transform(lambda x: x.cov(data['Profit']))  
data2=data[['Retail Sales People', 'Discount_Profit_Correlation', 'Discount_Profit_Covariance']].drop_duplicates()
data2
```

```python
#identifying top and bottom performers
top_performers=data2.sort_values(by='Discount_Profit_Correlation', ascending=False)
bottom_performers=data2.sort_values(by='Discount_Profit_Correlation')
print("Top Performers:\n", top_performers)
print("Bottom Performers:\n", bottom_performers)
```

```python
#5.group data by region and compute mean and variance of sales at different levels of discount(0%,1-10%,>10%), then visualize the results using box plots and histograms to compare distribution of sales across different discount levels.
# defining discount levels
def discount_level(discount):
    if discount == 0:
        return '0%'
    elif 0 < discount <= 0.1:
        return '1-10%'
    else:
        return '>10%'
data['Discount_Level'] = data['Discount'].apply(discount_level)

# Grouping by Region and Discount Level, then calculate mean and variance of Sales
grouped_data = data.groupby(['Region', 'Discount_Level'])['Sales'].agg(['mean', 'var']).reset_index()
print(grouped_data)

# Visualizing the results using box plots and histograms
plt.figure(figsize=(10,6))
sns.boxplot(x='Discount_Level', y='Sales', hue='Region', data=data)
plt.title('Boxplot of Sales by Discount Level and Region')      
plt.xlabel('Discount Level')
plt.ylabel('Sales')     
plt.legend(title='Region')
plt.show()

#visualizing the distribution of sales across different discount levels using histograms
plt.figure(figsize=(12, 6))
for discount_level in data['Discount_Level'].unique():
    sns.histplot(data[data['Discount_Level'] == discount_level]['Sales'], bins=30, kde=True, label=discount_level)
    plt.title('Histogram of Sales by Discount Level')
    plt.xlabel('Sales')
    plt.ylabel('Frequency')
    plt.legend(title='Discount Level')
    plt.show()
```

```python
#7.using the data, identify the top 5 regions with the highest sales and visualize their sales trends over time.
# Identify top 5 regions with highest sales
top_regions = data.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(5)
print("Top 5 Regions by Sales:\n", top_regions)

# Visualize sales trends of top regions over time
top_region_names = top_regions.index.tolist()
plt.figure(figsize=(12, 6))
for region in top_region_names:
    region_data = data[data['Region'] == region]
    sns.lineplot(x='Order Date', y='Sales', data=region_data, label=region)
    plt.title('Sales Trends of Top 5 Regions')
    plt.xlabel('Order Date')
    plt.ylabel('Sales')
    plt.legend(title='Region')
    plt.show()
```

```python
#6.using the data, identify the top 5 products with the highest profit margins and visualize their sales trends over time.
# Calculate profit margin for each product
data['Profit_Margin'] = data['Profit'] / data['Sales']  

# Group by Product Name and calculate mean profit margin
top_products = data.groupby('Product Name')['Profit_Margin'].mean().sort_values(ascending   =False).head(5)
print("Top 5 Products by Profit Margin:\n", top_products)   

# Visualize sales trends of top products over time
top_product_names = top_products.index.tolist()
plt.figure(figsize=(12, 6)) 
for product in top_product_names:
    product_data = data[data['Product Name'] == product]
    sns.lineplot(x='Order Date', y='Sales', data=product_data, label=product)   
    plt.title('Sales Trends of Top 5 Products by Profit Margin')
    plt.xlabel('Order Date')
    plt.ylabel('Sales')
    plt.legend(title='Product Name')    
    plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# --- Create the dataset ---
data = {
    'CustomerID': range(1, 16),
    'Annual_Income': [45,48,51,46,47,49,50,300,44,47,48,46,45,420,43],
    'Spending_Score': [42,39,44,46,41,40,43,90,45,43,42,44,43,5,41]
}
df = pd.DataFrame(data)

# --- Train Isolation Forest ---
iso = IsolationForest(contamination=0.1, random_state=42)
df['Outlier'] = iso.fit_predict(df[['Annual_Income', 'Spending_Score']])

# --- Map results for clarity ---
df['Outlier'] = df['Outlier'].map({1: 'Normal', -1: 'Outlier'})

# --- Visualize ---
plt.scatter(df['Annual_Income'], df['Spending_Score'], 
            c=(df['Outlier'] == 'Outlier'), cmap='coolwarm', s=80)
plt.xlabel('Annual Income (₹k)')
plt.ylabel('Spending Score')
plt.title('Outlier Detection using Isolation Forest')
plt.show()

# --- Display result ---
print(df)
```
